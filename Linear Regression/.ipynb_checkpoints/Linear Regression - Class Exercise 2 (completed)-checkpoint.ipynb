### Build a linear regression model and make the predcition on the same set
As we are making the prediction on the same set, we are only evaluating the goodness of fit.<br>

If we want to evaluate the model performance, we should split our data into training set and test set, build the model using training set and make prediction on the test set.

<font color=red><b>Action 1</b>: Initialize a linear regression model<br>
<b>Action 2</b>: Train the model by the train features and the train label<br>

### Record the metrics
We can create a list to store the evaluation metrics for future comparison.

```python
class performance_recorder():
    def __init__(self, y):
        self.performance_record = []
        self.y = y
        
    def add_new_record(self, name, yhat):
        r2 = metrics.r2_score(self.y, yhat)
        MSE = metrics.mean_squared_error(self.y, yhat)
        MAE = metrics.mean_absolute_error(self.y, yhat)
        self.performance_record.append([name, r2, MSE, MAE])
        
    def display(self):
        columns=['model', 'R2', 'MSE', 'MAE']
        display(pd.DataFrame(self.performance_record, columns=columns))
```

```python
recorder = performance_recorder(test_y)
recorder.add_new_record('LR', test_yhat)
recorder.display()
```

Take note that, we can display the record as a DataFrame object. But keep the record itself as a list for the ease of adding rows.

# Formulate Interaction
To formulate the interaction between X1 and X2, we can simply perform X1 * X2 and store it as a new feature

### Create a list of features to formulate interaction
The interaction term will be made between every pair of features in the list.

<font color=red><b>Action</b>: Create a list of variables to formulate the interaction between every pair.

```python
interaction_features = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research']
```

### Create a list of pairs
Using nested for loops, we can create a list of feature pairs.<br>

We have 7 features above, so there should be 21 unique combinations (regardless the sequence).

```python
feature_pairs = []
for i in range(len(interaction_features)):
    for j in range(i+1, len(interaction_features)):
        feature1 = interaction_features[i]
        feature2 = interaction_features[j]
        feature_pairs.append([feature1, feature2])
```

<font color=red><b>Action</b>: Determine the number of interaction terms below and verify if the number tallies with the length of the feature pairs.

```python
n_pairs = 21
print(n_pairs == len(feature_pairs))
```

```python
feature_pairs
```

### Formulate interaction terms and add to the DataFrame

```python
for feature1, feature2 in feature_pairs:
    df['{}_{}'.format(feature1, feature2)] = df[feature1] * df[feature2]

df
```

# Perform Feature Selection by p-values

### Extract the second table from Regression Report
This table is a DataFrame object.

```python
pvar_df = model.summary2().tables[1]
pvar_df
```

### Sort the DataFrame by ascending p-value
Now we are looking at this DataFrame from the most significant predictors to the least significant ones.

```python
pvar_df = pvar_df.sort_values(['P>|t|'], ascending=True).reset_index()
pvar_df
```

### Extract the ranked list of features
Remember to remove "const" from the list as it is not a feature.

```python
ranked_features = pvar_df['index'].to_list()
ranked_features.remove('const')
ranked_features
```

### Evaluate the model trained by the best k features (k = 1, 2, 3, ...)

This operation follows the procedures below:

    Loop from k = 1 to K
    In each loop,
        1) Use top k features by the lowest p-value as the feature set
        2) Rebuild the model
        3) Predict the test label
        4) Compute the performance metrics
        5) Add the performance metrics to the record

<font color=red><b>Action</b>: Complete the following cell.

```python
# Create an empty list as the record
feature_selection_record = []

# Loop over the number of features
for i in range(len(ranked_features)):
    # Set the number of features, k
    n_features = i + 1
    
    # Extract the first k features in the ranked list
    feature_set = ranked_features[:n_features]
    
    # Preparet the feature set and the label
    train_x = train_df[feature_set]
    test_x = test_df[feature_set]
    
    # Initialize and train the model
    model = LinearRegression()
    model.fit(train_x, train_y)
    
    # Make the prediction on the same set
    test_yhat = model.predict(test_x)
    
    # Determine MSE
    MSE = metrics.mean_squared_error(test_y, test_yhat)
    
    # Record the result
    feature_selection_record.append([n_features, MSE])
```

### Preview the record
This record shows MSE for the number of features we used.<br>
We should expect MSE to be decreasing as more features will result in a more complex model, hence a better fit on the training set.

```python
feature_selection_record = pd.DataFrame(feature_selection_record, columns=['n_features', 'MSE'])
feature_selection_record.head(10)
```

### Create a line chart for MSE vs the number of features used

```python
plt.figure(figsize=(12, 6))
plt.plot(feature_selection_record['n_features'], feature_selection_record['MSE'], marker='x')
plt.show()
```

We can observe that, MSE decreases dramatically in the beginning but get flat from k = 3 onwards.<br>
But let's play safe. We will take the best 5 features to build the model.

### Display the best 3 features and the corresponding MSE

```python
n_features_to_use = 5
feature_set = ranked_features[:n_features_to_use]
feature_set
```

### Re-run the model with the 5 best features selected by p-values and record the metrics

<font color=red><b>Action</b>: Rebuild the model and make the prediction on the test label

```python
train_x = train_df[feature_set]
test_x = test_df[feature_set]

model3 = LinearRegression()
model3.fit(train_x, train_y)
test_yhat = model3.predict(test_x)
```

```python
recorder.add_new_record('LR with interaction (best 5 by p-value)', test_yhat)
recorder.display()
```

# Perform Forward Feature Selection
The forward feature selection follows the procedures below:

    1) Initialize an empty list that will contain the selected features
    2) Take the features not selected yet as the feature candidate
    3) Loop over each feature candidate. In each loop,
        3.1) Use the selected features plus the feature candidate in the loop as the feature set
        3.2) Rebuild the model
        3.3) Predict the test label
        3.4) Compute the performance metrics
        3.5) Add the performance metrics to the record
    4) Identify the best feature candidate corresponding to the best performance
    5) Add the best feature candidate to the list of selected features
    6) If the number of selected features is less than the requirement, go back to Step 2; Else, over.

<font color=red><b>Action</b>: Complete the following cell.

```python
# Create a list to record the performance
forward_selection_record = []

# Create a list of selected features, it should be empty at first
selected_features = []

# Create a list of feature candidates, it should have all features at first
feature_candidates = features

# Loop until we reach the number of feature we want to use
# n_features_used has been defined as 6
for i in range(n_features_to_use):
    
    # Create a list to record MSE for this iteration
    MSE_record = []
    
    # Loop over all feature candidates
    for feature in feature_candidates:
        
        # The feature set should include the selected features and the feature under evaluation
        feature_set = selected_features + [feature]
        
        # Prepare the feature set and the label
        
        train_x = train_df[feature_set]
        test_x = test_df[feature_set]
        
        # Train the model and make the prediction
        model = LinearRegression()
        model.fit(train_x, train_y)
        test_yhat = model.predict(test_x)
        
        # Determine MSE
        MSE = metrics.mean_squared_error(test_y, test_yhat)
        
        # Append the result into forward selection record
        forward_selection_record.append([len(feature_set), feature_set, MSE])
        
        # Append the MSE into the record
        MSE_record.append(MSE)
    
    # After all feature candidates are evaluated
    
    # Get the minimum MSE for this iteration
    min_MSE = min(MSE_record)
    
    # Get the index of the minimum MSE
    index = MSE_record.index(min_MSE)
    
    # Get the feature candidate that gives the minimum MSE
    best_feature_to_add = feature_candidates[index]
    
    # Append the best feature into the list of selected features
    selected_features.append(best_feature_to_add)
    
    # Remove the best feature from the list of feature candidates
    feature_candidates.remove(best_feature_to_add)
    
# After wo reach the number of features

# Display the forward selection record
forward_selection_record = pd.DataFrame(forward_selection_record, columns=['n_features', 'feature_set', 'MSE'])
forward_selection_record
```

This DataFrame records all the attempts that we have done.<br>
We can find out the best feature set (that gives minimum MSE) for n_features == 5.

<font color=red><b>Action</b>: Filter the DataFrame to keep the rows with n_features == 5

```python
forward_result = forward_selection_record[forward_selection_record['n_features'] == n_features_to_use]
forward_result.head()
```

<font color=red><b>Action</b>: Sort the DataFrame by MSE in ascending order

```python
forward_result = forward_result.sort_values(['MSE'], ascending=True)
forward_result.head()
```

<font color=red><b>Action</b>: Obtain the feature set with the lowest MSE

```python
feature_set = forward_result['feature_set'].iloc[0]
feature_set
```

### Re-run the model with the 5 best features selected by <u>forward selection</u> and record the metrics

<font color=red><b>Action</b>: Rebuild the model and make the prediction on the test label

```python
train_x = train_df[feature_set]
test_x = test_df[feature_set]

model4 = LinearRegression()
model4.fit(train_x, train_y)
test_yhat = model4.predict(test_x)
```

```python
recorder.add_new_record('LR with interaction (best 5 by forward selection)', test_yhat)
recorder.display()
```

# Perform Backward Feature Selection
The backward feature selection follows the procedures below:

    1) Initialize a list that contains all features
    2) Loop over each selected features. In each loop,
        2.1) Use all selected features except the feature in the loop as the feature set
        2.2) Rebuild the model
        2.3) Predict the test label
        2.4) Compute the performance metrics
        2.5) Add the performance metrics to the record
    4) Identify the best feature to drop corresponding to the best performance
    5) Remove the best feature to drop from the list of selected features
    6) If the number of selected features is more than the requirement, go back to Step 2; Else, over.

<font color=red><b>Action</b>: Complete the following cell.

```python
# Create a list to record the performance
backward_selection_record = []

# Create a list of selected features, it should be empty at first
selected_features = features.copy()

# Loop until we reach the number of feature we want to use
# n_features_used has been defined as 6
for i in range(len(selected_features)):
    
    # Create a list to record MSE for this iteration
    MSE_record = []
    
    # Loop over all feature candidates
    for feature in selected_features:
        
        # The feature set should include the selected features and the feature under evaluation
        feature_set = selected_features.copy()
        feature_set.remove(feature)
        
        # Prepare the feature set and the label
        train_x = train_df[feature_set]
        test_x = test_df[feature_set]
        
        # Train the model and make the prediction
        model = LinearRegression()
        model.fit(train_x, train_y)
        test_yhat = model.predict(test_x)
        
        # Determine MSE
        MSE = metrics.mean_squared_error(test_y, test_yhat)
        
        # Append the result into backward selection record
        backward_selection_record.append([len(feature_set), feature_set, MSE])
        
        # Append the MSE into the record
        MSE_record.append(MSE)
    
    # After all features are excluded once
    
    # Get the minimum MSE for this iteration
    min_MSE = min(MSE_record)
    
    # Get the index of the minimum MSE
    index = MSE_record.index(min_MSE)
    
    # Get the feature candidate that gives the minimum MSE
    best_feature_to_drop = selected_features[index]
    
    # Remove the best feature from the list of feature candidates
    selected_features.remove(best_feature_to_drop)
    
    # Break the loop when there are only 6 features left
    if len(selected_features) == n_features_to_use:
        break
    
# After wo reach the number of features

# Display the forward selection record
backward_selection_record = pd.DataFrame(backward_selection_record, columns=['n_features', 'feature_set', 'MSE'])
backward_selection_record
```

This DataFrame records all the attempts that we have done.<br>
We can find out the best feature set (that gives minimum MSE) for n_features == 5.

<font color=red><b>Action</b>: Filter the DataFrame to keep the rows with n_features == 5 and sort the DataFrame by MSE in ascending order

```python
backward_result = backward_selection_record[backward_selection_record['n_features'] == n_features_to_use]
backward_result = backward_result.sort_values(['MSE'], ascending=True)
backward_result
```

<font color=red><b>Action</b>: Obtain the feature set with the lowest MSE

```python
feature_set = backward_result['feature_set'].iloc[0]
feature_set
```

### Re-run the model with the 5 best features selected by <u>backward selection</u> and record the metrics

<font color=red><b>Action</b>: Rebuild the model and make the prediction on the test label

```python
train_x = train_df[feature_set]
test_x = test_df[feature_set]

model5 = LinearRegression()
model5.fit(train_x, train_y)
test_yhat = model5.predict(test_x)
```

```python
recorder.add_new_record('LR with interaction (best 5 by backward selection)', test_yhat)
recorder.display()
```
